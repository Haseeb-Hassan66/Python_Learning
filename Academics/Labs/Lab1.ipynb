```python
# Task 2
cities = ["Karachi", "Lahore", "Islamabad", "Quetta", "Peshawar"]
countries = ("Pakistan", "India", "Sri Lanka", "Nepal", "Bangladesh")
languages = {"English", "Urdu", "Hindi", "Sindhi", "Punjabi"}

cities.append("Faisalabad")

country_selected = countries[2]

languages.add("Arabic")

cities_set = set(cities)

languages_list = list(languages)

print("Updated List of Cities:", cities)
print("Selected Country from Tuple:", country_selected)
print("Updated Set of Languages:", languages)
print("Cities as Set:", cities_set)
print("Languages as List:", languages_list)

# List is used when you need an ordered, changeable collection.
# Tuple is used when you need an ordered, immutable collection.
# Set is used when you need an unordered collection of unique items.
```

```python
# Task 3
Student_name = "Haseeb"
roll_number = "023-24-0274"
percentage = "80%"

marks = [80, 70, 60, 69, 91]
marks_tuple = (80, 70, 60, 69, 91)
marks_set = set(marks)

total = sum(marks)
average = total / 5

print("Total Marks:", total)
print("Average Marks:", average)

if 69 in marks:
    print("69 marks exists")
else:
    print("69 marks does not exist")

marks_set.add(100)

print("\nStudent Name:", Student_name)
print("Roll Number:", roll_number)
print("Percentage:", percentage)
print("Marks List:", marks)
print("Marks Tuple:", marks_tuple)
print("Marks Set:", marks_set)

# A set may contain fewer values than a list because sets store only unique items, removing duplicates automatically.
```
