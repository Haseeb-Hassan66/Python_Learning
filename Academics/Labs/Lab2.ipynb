```python
# Task 1
Students = {
    "Taraz": [60, 68, 70],
    "Haseeb": [76, 81, 85],
    "Akbar": [71, 70, 75],
    "Saba": [60, 86, 80],
    "Anees": [87, 66, 90]
}

top_performer = None
top_marks = 0

for name in Students:
    total_marks = 0

    for mark in Students[name]:
        total_marks += mark

    avg_marks = total_marks / len(Students[name])

    if avg_marks >= 80:
        print("Average marks of", name, "is", avg_marks," and His Grade is A")
    elif avg_marks >= 60:
        print("Average marks of", name, "is", avg_marks, " and His Grade is B")
    elif avg_marks >= 40:
        print("Average marks of", name, "is", avg_marks, " and His Grade is C")
    else:
        print("Average marks of", name, "is", avg_marks, " and His Grade is F")

    if avg_marks > top_marks:
        top_marks = avg_marks
        top_performer = name

print("Top performer is", top_performer, "with", top_marks, "marks")
```

```python
# Task 3
import numpy as np

sensor_readings = np.array([-1, 101, 50, 25, 0])

print("Invalid readings:", end=" ")
for value in sensor_readings:
    if value < 0 or value > 100:
        print(value, end=" ")
print()

total = 0
valid_count = 0

for value in sensor_readings:
    if value >= 0 and value <= 100:
        total += value
        valid_count += 1

average = total / valid_count

corrected_count = 0

for i in range(len(sensor_readings)):
    if sensor_readings[i] < 0 or sensor_readings[i] > 100:
        sensor_readings[i] = average
        corrected_count += 1

print("Updated sensor readings:", sensor_readings)
print("Number of readings corrected:", corrected_count)
```
